```python
#read input file
#fin = open("/Users/jannylund/bbk/first.txt", "rt")
#fout = open("/Users/jannylund/bbk/firstout.txt", "wt")
#fin.close()
#fout.close()
fin = open(r"C:\Users\nylund\Onedrive-Privat\OneDrive\jannesgit\python\bbk\bbk\telia-out.txt", "rt")
fout = open(r"C:\Users\nylund\Onedrive-Privat\OneDrive\jannesgit\python\bbk\bbk\out-telia-cleaned.txt", "wt")

newdata=""
newline=""
#for line in fin:
#    print(line)
```

```python
for line in fin:
    #print(line)
    
    #read replace the string and add to newdata string
    if line.find('BST')!=-1:
        data = (line[len(line)-(len(line)-4):])
        data = (data[:len(data)-10])
        newline = newline + '2021-05-'
        newline = newline + data[:2] + ' '
        print(data)
        if data.find('PM')!=-1:
            hour=12
            #print(hour)
            if data[12:14].find('12')!=1:
                hour=int(data[12:14])+12
                #print(str(hour) + 'i loop')
        else:
            hour=int(data[7:9])
        if hour < 10:
            newline = newline + '0'
        newline = newline + str(hour)
        newline = newline + data[9:15]
        data = "'" + newline + "',"
        #print(data + ' ' + newline + str(hour))
        newline=""
        #data = (line.replace("BST\n", "BST',"))
        #data = (data.replace(' ', '-'))
        #data = "'" + data
    else:
        data = (line.replace("BST\n", "BST',"))
        data = (data.replace(' ', ','))
        data = (data.replace('sth', "'sth"))
        data = (data.replace('\n', "'\n"))
        
    newdata=newdata + data
    
```

```python
#open the input file in write mode
#fin = open("data.txt", "wt")
#overrite the input file with the resulting data
fout.write(newdata)
#close the file
fin.close()
fout.close()
```
